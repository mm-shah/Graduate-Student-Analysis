## Men Vs Women in Major Category

```python
major_men_dict = {}
major_women_dict = {}
for i in range(len(recent_grads)):
    if i == 21:
        continue
    else:
        category = recent_grads['Major_category'][i]
    if category in major_men_dict:
        major_men_dict[category]+=recent_grads['Men'][i]
        major_women_dict[category]+=recent_grads['Women'][i]
    else:
        major_men_dict[category]=recent_grads['Men'][i]
        major_women_dict[category]=recent_grads['Women'][i]
major_men_dict_len = len(major_men_dict)
major_women_dict_len = len(major_women_dict)
```

```python
fig = plt.figure(figsize=(8,8))
plt.bar(np.arange(major_men_dict_len)-0.3, list(major_men_dict.values()),width=0.3, color= 'b', align= 'center')
plt.bar(np.arange(major_women_dict_len), list(major_women_dict.values()),width=0.3, color= 'r', align='center')
plt.xlabel('Major Category')
plt.ylabel('# of People')
plt.legend(('Men','Women'))
plt.xticks(np.arange(major_women_dict_len)-0.15,list(major_men_dict.keys()), rotation = 90)
plt.show()
```
